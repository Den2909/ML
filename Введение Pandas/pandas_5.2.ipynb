```python
import pandas as pd
melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head()
```

```python
melb_df.info()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
melb_df['quart_sell'] = melb_df['Date'].dt.quarter
melb_df['quart_sell'].value_counts()
```

```python
melb_df.nunique()
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] 
max_unique_count = 150 
for col in melb_df.columns: 
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: 
        melb_df[col] = melb_df[col].astype('category')
display(melb_df.info())
```

```python
melb_df.sort_values(by=['Price', 'Distance']).loc[::10, ['Distance', 'Price']]
```

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
melb_df.sort_values(
    by='AreaRatio', 
    ignore_index=True,
    ascending=False
).loc[1558, 'BuildingArea']
```

```python
mask1 = melb_df['Type'] == 'townhouse'
mask2 = melb_df['Rooms'] > 2
int(melb_df[mask1&mask2].sort_values(
    by=['Rooms', 'MeanRoomsSquare'],
    ascending=[True, False],
    ignore_index=True
).loc[18, 'Price'])
```

```python
mask1 = melb_df['Type'] == 'townhouse'
mask2 = melb_df['Rooms'] > 2
melb_df_room = melb_df[mask1 & mask2].sort_values(
    by='Rooms'
)
melb_df_room.sort_values(
    by='MeanRoomsSquare',
    ignore_index=True,
    ascending=False
).loc[18, 'Price']
```
